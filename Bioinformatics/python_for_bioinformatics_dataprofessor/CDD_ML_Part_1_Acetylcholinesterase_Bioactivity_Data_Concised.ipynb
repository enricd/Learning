### ChEMBL Database

The [ChEMBL Database](https://www.ebi.ac.uk/chembl/) is a database that contains curated bioactivity data of more than 2 million compounds. It is compiled from more than 76,000 documents, 1.2 million assays and the data spans 13,000 targets and 1,800 cells and 33,000 indications. [Data as of March 25, 2020; ChEMBL version 26].

```python
# Import libraries

import pandas as pd
from chembl_webresource_client.new_client import new_client
```

### Search for Target protein

```python
# Target search for coronavirus

target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase 

selected_target = targets.target_chembl_id[6]
selected_target
```

```python
# Here, we will retrieve only bioactivity data for coronavirus 3C-like proteinase (CHEMBL3927) that are reported as IC50 values.

activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.standard_type.unique()
```

```python
df.to_csv("data/bioactivity_data.csv", index=False)
```

```python
df = pd.read_csv("data/bioactivity_data.csv")
```

```python
df2 = df[df.standard_value.notna()]
```

### Data pre-processing of the bioactivity data

```python
# Labeling compounds as either being active, inactive or intermediate

# The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered
# to be active while those greater than 10000 nM will be considered to be inactive. As for those values in between
# 1000 and 10000 nM will be referred to as intermidiate

bioactivity_class = []
for v in df2.standard_value:
    if float(v) >= 10000:
        bioactivity_class.append("inactive")
    elif float(v) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

```python
# Iterate the molecule_chembl_id to a list

mol_cid = []
for i in df2.molecule_chembl_id:
    mol_cid.append(i)

mol_cid
```

```python
# Iterate canonical_smiles to a list

canonical_smiles = []
for i in df2.canonical_smiles:
    canonical_smiles.append(i)

canonical_smiles
```

```python
selection = ["molecule_chembl_id", "canonical_smiles", "standard_value"]
df3 = df2[selection]

df3
```

```python
df4 = pd.concat([df3, pd.Series(bioactivity_class)], axis=1)
df4
```

```python
df4.to_csv("data/bioactivity_preprocessed_data.csv", index=False)
```
